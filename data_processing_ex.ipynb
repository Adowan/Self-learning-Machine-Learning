## Making Blob

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
style.use( 'fivethirtyeight')
X,y= make_blobs(n_samples=500, centers= 3, cluster_std=1, n_features=2)
plt.scatter(X[:,0], X[:,1], s= 10, color='b')
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
plt.clf()
```

## Creating Moon

```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
from matplotlib import style
style.use("fivethirtyeight")
X,y = make_moons( n_samples = 500, noise = 0.1)
plt.scatter( X[ :, 0], X[ :,1], s=10 , color= 'r')
plt.xlabel("Horizontal Space")
plt.ylabel(" Vertical Space")
plt.figure(figsize=(20,10), dpi= 300)
plt.show()
```

## Now creating Dataset for ML

```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
from matplotlib import style
X,y = make_moons ( n_samples= 500, shuffle= True, noise= 0.15, random_state= 42)
plt.scatter( X[:, 0], X[:, 1], s=15, c=y)
plt.xlabel("Horizontal Space")
plt.ylabel(" Vertical Space")
plt.figure( figsize=(20,10), dpi= 300)
plt.show()
 
```

```python

 
from sklearn.datasets import make_multilabel_classification
import pandas as pd
import matplotlib.pyplot as plt
 
 
X, y = make_multilabel_classification(n_samples=500, n_features=2,
                                      n_classes=2, n_labels=2,
                                      allow_unlabeled=True,
                                      random_state=23)
 
df = pd.concat([pd.DataFrame(X, columns=['X1', 'X2']),
                pd.DataFrame(y, columns=['Label1', 'Label2'])],
               axis=1)
display(df.head(10))
 
plt.figure( figsize=(10,5), dpi= 300)
plt.scatter(df['X1'], df['X2'],s=10, c=df['Label1'])
plt.show()
```

```python

 
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
 
X, y = make_regression(n_samples = 300, n_features=1,noise=20, random_state=23)
plt.figure(figsize=(20,15), dpi=300) 
plt.scatter(X, y, s= 15, c=y)
plt.show()
```

```python
from sklearn.datasets import make_sparse_uncorrelated
import matplotlib.pyplot as plt
 
X, y = make_sparse_uncorrelated(n_samples = 100, n_features=4, random_state=23)
 
plt.figure(figsize=(12,10), dpi=300)
for i in range(4):
    plt.subplot(2,2, i+1)
    plt.scatter(X[:,i], y, s=10)
    plt.xlabel('X'+str(i+1))
    plt.ylabel('Y')
plt.show()
```
